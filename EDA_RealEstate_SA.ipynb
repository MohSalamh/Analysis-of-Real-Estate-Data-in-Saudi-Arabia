##### **Mapping Rent Period ID to its Label**

**This from the dataset**
rent_period: integer, id of a period (0: Yearly | 1: Daily | 2: Monthly | 3: Yearly)

```python
# define a list to map rent periods
rentPeriod_txt_id = ['Yearly', 'Daily', 'Monthly', 'Sale']

# Function to convert rent period identifier to text
def rent_period_to_text(category, rent_period):
    if "Sell" in category:
        return rentPeriod_txt_id[3]
    elif rent_period == 0 or rent_period == 3: # Treating 3 as Yearly as per the original structure
        return rentPeriod_txt_id[0]
    elif rent_period == 1:
        return rentPeriod_txt_id[1]
    elif rent_period == 2:
        return rentPeriod_txt_id[2]
```

### **8. Counts of Listings by Different Dimensions**

##### **Functions**
We will write three funcitons:<br>
1. **get_count_stats** func which takes two args namely the dataframe and the name of a categorical column and return dataframe containing the count and percentage of each class of that categorical column.
2. **consolidate_category** func which three args namely the dataframe, the column name for which to reduce the number of classes and an integer that represents the number of classes check (if more consolidate the classes and if not leave as such). Basically, the bottom **<span style="color: red;">5%</span>** of classes by counts will be consolidated in the generalized class "Others".<br>
3. **bar_plotting** func which takes three inputs namely the dataframe, the column name and a boolean indicating if it will be horizental or vertical bar plot. The func displays horizental bar chart showing the distribution of listings by each categorical field. 

```python
# generalized formula to help up retriving the counts by different dimensions quickly
def get_count_stats(df, col_name):
    # get counts
    count_ser = df[col_name].value_counts()
    # find the total count of the records
    tot_count = count_ser.sum()
    
    # create list of enteries for each field
    dim_name = list(count_ser.index.values)
    num_byDim = list(count_ser)
    # find the percentage of each class
    per_byDim = [i/tot_count*100 for i in list(count_ser)]
    
    # Creating a dictionary from lists
    data = {
        col_name :  dim_name,
        'count': num_byDim,
        'percentage': per_byDim
    }
    
    # Creating DataFrame
    df_data = pd.DataFrame(data)

    # make sure categorical fields are transformed to categorical field in the new dataframe
    if df[col_name].dtype == 'category':
        category_order = list(df[col_name].unique().sort_values())
        df_data[col_name] = pd.Categorical(df_data[col_name], categories=category_order)
    
    return df_data
```

```python
def consolidate_category(df, col_name, max_cat):
    if df[col_name].nunique() > max_cat:
        df_agg = get_count_stats(df, col_name)
        # create a column that calcualte the cumeraltive percentage
        df_agg['cumulative_percetage'] = df_agg[df_agg.columns[2]].cumsum()
        # classify cities that are the bottom 5% of the cohort as 'others'
        df_agg[df_agg.columns[0]] = df_agg[df_agg.columns[0]].where(df_agg['cumulative_percetage'] <= 95, 'Others')
        # create new df that groups the others and other cities by count 
        df_agg = df_agg.groupby(df_agg.columns[0])[[df_agg.columns[1], df_agg.columns[2]]].sum()
        df_agg = df_agg.sort_values(by=df_agg.columns[1], ascending=False).reset_index()
        return df_agg
    else:
        return df.copy()
```

```python
# import matplotlib.ticker as ticker
import math

# define a function to plot the a histogram
def bar_plotting(df, col_name, horizental_check):
    
    # determine label offsets
    if df['count'].max() <= 13594:
        label_offset1 = 10
    elif df['count'].max() >= 13594 & df['count'].max() <= 201071:
        label_offset1 = (df['count'].max() - 9584.4)/ 190.62
    else:
        label_offset1 = 1000
        
    label_offset2 = (df['count'].max() - 937.59)/ 11.612

    if col_name in ['month', 'week_day']:
        df_sorted = df.sort_values(by=col_name, ascending=True)
    else:
        df_sorted = df.sort_values(by='count', ascending=True)
        
    if 'Others' in df_sorted[col_name].values:
        # Specify the category to move to the bottom
        move_to_bottom = 'Others' 
        # Reorder categories based on the condition
        category_order = [move_to_bottom] + [cat for cat in df_sorted[col_name] if cat != move_to_bottom]
        df_sorted[col_name] = pd.Categorical(df_sorted[col_name], categories=category_order)
        df_sorted = df_sorted.sort_values(by=col_name, ascending=True)
        
    values = list(df_sorted['count'])
    # Use a colormap for the bar colors
    cmap = plt.cm.cividis
    norm = plt.Normalize(min(values), max(values))  # Normalize the values for color mapping

    if horizental_check:
        ax = df_sorted['count'].plot(kind='barh', color=cmap(norm(values)), figsize=(10, 6))
        ax.set_yticklabels(df_sorted[col_name])
        ax.set_xticklabels([])
    else:
        ax = df_sorted['count'].plot(kind='bar', color=cmap(norm(values)), figsize=(10, 5.5))
        ax.set_yticklabels([])
        ax.set_xticklabels(df_sorted[col_name], rotation=0)

    # Set the background for the plot area
    ax.set_facecolor('white')

    # remove characters from the title
    title = col_name.replace("_en", "").replace("_", " ").strip()
    
    #Adding title 
    plt.title('Number of Listings Created by ' + title.title())
    
    # disable the grid
    plt.grid(False)

    bar_width = ax.patches[0].get_width()
    
    # Add labels to the bars
    for i, (v, p) in enumerate(zip(df_sorted['count'], df_sorted['percentage'])):
        if horizental_check:
            # labels for horizental bars
            ax.text(v+label_offset1, i, '{:,.0f}'.format(v), color='black', va='center', fontsize=11)  
            ax.text(v+label_offset2, i, '{:.1f}%'.format(p), color='black', va='center', fontsize=11,
               bbox=dict(facecolor='lightblue', boxstyle='round,pad=0.2'))
        else:
            # labels for vertical bars
            ax.text(i-bar_width/2, v+label_offset2/3, '{:,.0f}'.format(v), color='black', va='center', fontsize=11)  
            ax.text(i-bar_width/2, v+label_offset2, '{:.1f}%'.format(p), color='black', va='center', fontsize=11,
               bbox=dict(facecolor='lightblue', boxstyle='round,pad=0.2'))

    # Display the plot
    plt.show()
```

In this section, we will write one funciton:<br>
**histo_plots** func which takes two args namely the dataframe and the field column name and generates two distribution plots: one for Rent and one for Sale properties.

```python
# Let's create function to show the distribution for variables for (Rent and Sale transactions)
def histo_plots(df, out_col):
    # Create subplots for side-by-side histograms
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    # Plot the first histogram
    axes[0].hist(df[(df['transaction_type'] == 'Rent')][out_col], bins=30, color='blue', alpha=0.7)
    axes[0].set_title(f'Distribution of {out_col.capitalize()} for Rent Properties')
    axes[0].set_xlabel('Value')
    axes[0].set_ylabel('Frequency')

    # Plot the second histogram
    axes[1].hist(df[(df['transaction_type'] == 'Sale')][out_col], bins=30, color='green', alpha=0.7)
    axes[1].set_title(f'Distribution of {out_col.capitalize()} for Sale Properties')
    axes[1].set_xlabel('Value')
    axes[1].set_ylabel('Frequency')

    # Adjust layout to avoid overlap
    plt.tight_layout()

    # Show the plots
    plt.show()
```

### **10. Bivariate Analysis**

##### **Functions**
We will write seven funcitons:<br>
1. **map_rent_type** func which takes a string input and match to a string sentence to be used for the corrsponding subplot tite.<br>
2. **consildated_list** func which calls the consolidated_category func and returns a list of the unique classes for the categorical column of interest,<br>
3. **remove_outliers** func which takes two args namely the dataframe, the numrical column name for which outliers will be identified and removed using **Interquartile Range**.<br>
4. **remove_low_count** func which takes two inputs namely the dataframe and the name of categorical column and remove classes **fewer than <span style="color: red;">10 </span>occurrences**. <br>
5. **stackedbarplots** func which takes two inputs: dataframe and categorical column name, and creates **Four plots** that shows the listings distribution for different rent periods namely (Sale, Annual, Montly and Daily).<br>
6. **avgPrice_barplots** func which takes two inputs: dataframe and categorical column name, and creates **Four plots** that shows the average price of Aparatment listings grouped by the cateogrical column accross different rent periods namely (Sale, Annual, Montly and Daily).<br>
7. **scatter_plots** func which takes one input: dataframe that contains Price and Area columns, and creates **Four scatter plots** of Average Price vs Average Area grouped by district ids for different rent periods namely (Sale, Annual, Montly and Daily).<br>

```python
# list of rent_types
rent_type = ['Sale', 'Yearly', 'Monthly', 'Daily']

# rect for main title
rect_title=[0, 0, 1, 0.96]
```

```python
# a function to map strings to subplot title
def map_rent_type(rent_type):
    match rent_type:
        case 'Sale':
            return "Properties for Sale"
        case 'Yearly':
            return "Properties for Annual Rent"
        case 'Monthly':
            return "Properties for Monthly Rent"
        case 'Daily':
            return "Properties for Daily Rent"
```

```python
# return a list of unique classes of category after consolidating classes if they exceed threshold
def consildated_list(df, cat, max_cat):
    return list(consolidate_category(df, cat, max_cat)[cat].unique())
```

```python
def remove_outliers(df, col):
    # Calculate Q1, Q3, and IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    # Define the bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove outliers
    return df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
def remove_low_count(df, cat):
    # Count occurrences of each category
    category_counts = df[cat].value_counts()

    # Filter out categories with a count lower than the threshold
    filtered_categories = category_counts[category_counts >= 10].index

    # Filter the original DataFrame to keep only these categories
    return df[df[cat].isin(filtered_categories)]
```

```python
def stackedbarplots(df, col_name):
    include_cat = consildated_list(df, col_name, 20)
    
    # Stacked bar plot for City vs Category
    # Create subplots for side-by-side histograms
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    
    for i, ax in enumerate(axes.flat):  
        df_local = df.copy()
        # filter dataframe
        df_local = df_local[df_local['rent_period'] == rent_type[i]]

        # consildate the classes by labeling the bottom 5% as others
        include_property = consildated_list(df_local, 'property_category', 1)
        df_local[col_name] = df_local[col_name].apply(lambda x: x if x in include_cat else 'Others')
        df_local['property_category'] = df_local['property_category'].apply(lambda x: x if x in include_property else 'Others')
        
        city_category = pd.crosstab(df_local[col_name], df_local['property_category'])

        # order the classes 
        custom_city_order = list(city_category.sum(axis=1).sort_values(ascending=True).index)
        if 'Others' in custom_city_order:
            custom_city_order.remove('Others')
            custom_city_order = ['Others'] + custom_city_order
        property_order = city_category.sum(axis=0).sort_values(ascending=False).index
        city_category = city_category.loc[custom_city_order, property_order]

        # creating the plot
        city_category.plot(kind='barh', stacked=True, ax=ax)
        ax.legend(loc='lower right')
        ax.set_title(f'{map_rent_type(rent_type[i])}')
        ax.tick_params(axis='y', labelsize=10)
        ax.set_xlabel('')  # Remove x-axis label
        ax.set_ylabel('')  # Remove y-axis label
        
     # Adjust layout to prevent overlap
    plt.tight_layout(rect=rect_title)  # Leave space for the main title

    fig.suptitle('Properties Ditribution by '+col_name.replace('_en', '').title(), fontsize=14)
    
    plt.show()
```

```python
# let's us plot the average price by city for apartments for different transaction types
def avgPrice_barplots(df, col_name):

    df_local = df.copy()
        
    # Create subplots for side-by-side histograms
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    
    for i, ax in enumerate(axes.flat):  # axs.flat flattens the 2x2 array to make it iterable
        # filter out irrelvant records
        df_filtered = df_local[(df_local['rent_period'] == rent_type[i]) & (df_local['property_category'] == 'Apartment')]
        df_filtered = remove_low_count(df_filtered, col_name)
        df_filtered = remove_outliers(df_filtered, 'price')

        # consildate classes
        include_list = consildated_list(df_filtered, col_name, 20)
        if 'Others' in include_list:
            df_filtered[col_name] = df_filtered[col_name].apply(lambda x: x if x in include_list else 'Others')

        # Aggregate by col_name
        df_grouping = df_filtered.groupby(col_name)['price'].mean().reset_index()
        df_grouping = df_grouping.sort_values(by='price', ascending=False)
        
        # Reorder categories based on the condition
        if 'Others' in df_grouping[col_name].unique():
            category_order = [cat for cat in df_grouping[col_name] if cat != 'Others'] + ['Others']
            df_grouping[col_name] = pd.Categorical(df_grouping[col_name], categories=category_order)

        # create the plot
        sns.barplot(data=df_grouping, y=col_name, x='price', hue='price', palette='viridis', orient='h', ax=ax)
        ax.set_title(f'{map_rent_type(rent_type[i])}')
        ax.get_legend().remove()
        ax.tick_params(axis='y', labelsize=10)
        ax.set_xlabel('')  # Remove x-axis label
        ax.set_ylabel('')  # Remove y-axis label
        
    # Adjust layout to prevent overlap
    plt.tight_layout(rect=rect_title)  # Leave space for the main title

    fig.suptitle('Apartment Properties: Avg Price vs '+col_name.replace('_en', '').title(), fontsize=14)
    
    # Show the plots
    plt.show()
```

```python
# Let's create function to plot scatter plot
def scatter_plots(df):
    
    # Create subplots for side-by-side histograms
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    
    for i, ax in enumerate(axes.flat):  # axs.flat flattens the 2x2 array to make it iterable
        # filter out irrelvant records (this is Riyadh dataframe)
        df_filtered = df[(df['city_en'] == 'Riyadh') & (df['rent_period'] == rent_type[i]) & (df['property_category'] == 'Apartment')]
        df_filtered = remove_low_count(df_filtered, 'district_id')
        df_filtered = remove_outliers(df_filtered, 'price')

        # Aggregate by Districts
        df_grouping = df_filtered.groupby('district_id')[['area', 'price']].mean()

        # filter out irrelvant records (this include all cities except for Riyadh)
        df_filtered_ = df[(df['city_en'] != 'Riyadh') & (df['rent_period'] == rent_type[i]) & (df['property_category'] == 'Apartment')]
        df_filtered_ = remove_low_count(df_filtered_, 'district_id')
        df_filtered_ = remove_outliers(df_filtered_, 'price')
        
        # Aggregate by Districts
        df_grouping_ = df_filtered_.groupby('district_id')[['area', 'price']].mean()

        # create scatter plot
        sns.scatterplot(x='area', y='price', data=df_grouping_, color='blue', label='Other Cities', alpha=0.7, ax=ax)
        sns.scatterplot(x='area', y='price', data=df_grouping, color='green', label='Riyadh', alpha=0.7, ax=ax)

        ax.set_title(f'{map_rent_type(rent_type[i])}')
        # show points for properties below 300 m2
        ax.set_xlim(0, 300)

    # Adjust layout to prevent overlap
    plt.tight_layout(rect=rect_title)  # Leave space for the main title

    fig.suptitle('Apartment Properties: Avg Price vs Area', fontsize=14)
    
    # Show the plots
    plt.show()
```

### **11. Correlation Matrices**

##### **Functions**
**correlation_matrices** func takes a dataframe as input and creates four correlation matrices of the variables (Area, Beds, Livings, WC, Street_Width, Age and Price ) across different rent periods.<br>

```python
def correlation_matrices(df):

    # Create subplots for side-by-side histograms
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    
    for i, ax in enumerate(axes.flat):  
        # Correlation heatmap to see correlations between numerical variables
        df_filtered = df[df['rent_period'] == rent_type[i]]
        df_filtered = remove_low_count(df_filtered, 'district_id')
        df_filtered = remove_outliers(df_filtered, 'price')
        correlation_matrix = df_filtered[['area', 'beds', 'livings', 'wc', 'street_width', 'age', 'price']].corr()
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, ax=ax)
        ax.set_title(f'{map_rent_type(rent_type[i])}')

    # Adjust layout to prevent overlap
    plt.tight_layout(rect=[0, 0, 1, 0.96])  # Leave space for the main title

    fig.suptitle('Correlation Matrices', fontsize=14)
    
    plt.show()
```
